```python
data1=df1['Gender'].unique().tolist()
data2=df1['Married'].unique().tolist()
data3=df1['Dependents'].unique().tolist()
data4=df1['Education'].unique().tolist()
data5=df1['Self_Employed'].unique().tolist()
```

```python
data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)
data2.replace({'Yes':'Married','No':'Not Married'},inplace=True)
data3 = pd.DataFrame(data3)
data3.replace({'0':'Dependents:0','1':'Dependents:1','2':'Dependents:2','3+':'Dependents:3+'},inplace=True)
data4 = pd.DataFrame(data4)
data5 = pd.DataFrame(data5)
data5.replace({'Yes':'Self Employed','No':'Not Self Employed'},inplace=True)
```

```python
combine = data1.append([data2,data3,data4,data5])
combine.rename(columns={0:'Variance'},inplace=True)
```

```python
combine = combine.dropna()
```

```python
counts_yes1 = df1[accepted]['Gender'].value_counts().tolist()
counts_no1 = df1[-accepted]['Gender'].value_counts().tolist()

counts_yes2 = df1[accepted]['Married'].value_counts().tolist()
counts_no2 = df1[-accepted]['Married'].value_counts().tolist()

counts_yes3 = df1[accepted]['Dependents'].value_counts().tolist()
counts_no3 = df1[-accepted]['Dependents'].value_counts().tolist()

counts_yes4 = df1[accepted]['Education'].value_counts().tolist()
counts_no4 = df1[-accepted]['Education'].value_counts().tolist()

counts_yes5 = df1[accepted]['Self_Employed'].value_counts().tolist()
counts_no5 = df1[-accepted]['Self_Employed'].value_counts().tolist()

result1 = (pd.DataFrame([counts_yes1,counts_no1])).transpose()

result2 = (pd.DataFrame([counts_yes2,counts_no2])).transpose()

result3 = (pd.DataFrame([counts_yes3,counts_no3])).transpose()

result4 = (pd.DataFrame([counts_yes4,counts_no4])).transpose()

result5 = (pd.DataFrame([counts_yes5,counts_no5])).transpose()

result = result1.append([result2,result3,result4,result5])
```

```python
new_data = pd.concat([combine,result],axis='columns',sort=False)
new_data = new_data.set_index('Variance')
new_data.rename(columns={0:'Accepted',1:'Not Accepted'},inplace=True)
new_data
```
